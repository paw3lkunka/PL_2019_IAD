```python
import csv
from math import sqrt as sqrt
from prettytable import PrettyTable
import matplotlib.pyplot as plot
import matplotlib.lines as mlines
from collections import defaultdict, Counter
import operator
from random import random

#######################################

path = input("Podaj ścieżkę do pliku TENINGOWEGO CSV:")
file = open(path)
reader = csv.reader(file)

header = next(reader)	# pierwszy rekord pliku uznaję za nagłówek do tabeli
types = next(reader)	# drugi rekord będzie opisywał typ każdej kolumny

def columnType(t,value):
	if t == 'float':
		return float(value)
	elif t == 'int':
		return float(value)
	elif t == 'string':
		return value
	elif t == 'flowerSpecies':
		switch = {'0':"setosa", '1':"versicolor", '2':"virginica"}
		return switch[value]
	else:
		return value 

def csv2matrix(reader):
	data = []
	for row in reader:
		tmp = []
		for index,val in enumerate(row):
			tmp.append(columnType(types[index],val))
		
		data.append(tmp)
	return data

def column2array(matrix, n):
	return [row[n] for row in matrix]

data = csv2matrix(reader)
file.close()
        
```

```python
def euclideanDistance(object1, object2, whichValues):
    distance = 0
    for val in whichValues:
        distance += pow(object1[val] - object2[val], 2)
    return sqrt(distance)

def randomPoint(xMin, xMax, yMin, yMax):
    return [(random() * (xMax - xMin)) + xMin, (random() * (yMax - yMin)) + yMin];

xMin = min(column2array(data, 0))
xMax = max(column2array(data, 0))
yMin = min(column2array(data, 1))
yMax = max(column2array(data, 1))
           
def generateCentroids(amount, xMin, xMax, yMin, yMax):
    centroids = []
    for i in range(amount):
        centroids.append(randomPoint(xMin, xMax, yMin, yMax))
    return centroids


def getPointsDistances(centroids, points):
    pointsDistances = []
    for point in points:
        pointDistances =[]
        for centroid in centroids:
            pointDistances.append(euclideanDistance(centroid, point, [0, 1]))
        pointsDistances.append(pointDistances)
    return pointsDistances

def getPointsAdherence(centroidDistances):
    pointsAdherence = []
    for point in centroidDistances:
        pointsAdherence.append(point.index(min(point)))
    return pointsAdherence

def centroidsMovement(centroids, points, pointsAdherence):
    segregatedPoints = []
    for centroid in centroids:
        segregatedPoints.append([])
    for index, adherence in enumerate(pointsAdherence):
        segregatedPoints[adherence].append(points[index])
    for index, it in enumerate(segregatedPoints):
        if len(it) == 0:
            pass
        else:
            centroids[index] = [sum(point[0] for point in it) / len(it), sum(point[1] for point in it) / len(it)]
    return centroids

def quantError(centroids, points, pointsAdherence):    
    quantError = 0.0
    for index, point in enumerate(points):
        quantError += pow(euclideanDistance(centroids[pointsAdherence[index]], point, [0, 1]), 2)

    return quantError / len(points)

def mergeQuantErrors(allQuantErrors, maxIterations):
    mergedErrors = []
    tmp = []
    for it in range(maxIterations):
        tmp.append([])
        for epoque in allQuantErrors:
            if(it < len(epoque)):
                tmp[it].append(epoque[it])
            else:
                tmp[it].append(epoque[len(epoque) - 1])
            
    for it in tmp:
        mergedErrors.append((sredniaArytmetyczna(it),sqrt(wariancjaSzcz(it))))
        
    return mergedErrors

def sredniaArytmetyczna(X):
    if not len(X) == 0:
        return sum(X) / len(X)
    else:
        return 0

def wariancjaSzcz(X):
    if not len(X) == 0:
        avr = sredniaArytmetyczna(X)
        VarX = (X[0] - avr)**2 
        for val in X[1:]:
            VarX += (val - avr)**2
        return VarX / len(X)
    else:
        return 0

amounts = [6, 12, 30]

iterationsStats = []
for amount in amounts:
    quantErrorsStats = []
    currentIterationsStats = []
    finalCentroids = []
    for it in range(12):
        centroids = generateCentroids(amount, xMin, xMax, yMin, yMax)
        pointsDistances = getPointsDistances(centroids, data)
        currentAdherence = getPointsAdherence(pointsDistances)
        iterationsAmount = 0
        previousAdherence = []
        currentQuantErrors = []

        while currentAdherence != previousAdherence and iterationsAmount < 200:
            iterationsAmount += 1
            centroids = centroidsMovement(centroids, data, currentAdherence)
            previousAdherence = currentAdherence
            pointsDistances = getPointsDistances(centroids, data)
            currentAdherence = getPointsAdherence(pointsDistances)
            currentQuantErrors.append(quantError(centroids, data, currentAdherence))

        quantErrorsStats.append((sredniaArytmetyczna(currentQuantErrors), sqrt(wariancjaSzcz(currentQuantErrors)), currentQuantErrors))
        currentIterationsStats.append(iterationsAmount)
        finalCentroids.append(centroids)
        print(centroids)
        
    iterationsStats.append((sredniaArytmetyczna(currentIterationsStats), sqrt(wariancjaSzcz(currentIterationsStats))))
    #print(quantErrorsStats)
    
    #worstQuantIndex = column2array(quantErrorsStats, 0).index(min(column2array(quantErrorsStats, 0)))
    #worstQuantErrors = quantErrorsStats[worstQuantIndex][2]
    allQuantErrors = column2array(quantErrorsStats, 2)
    mergedQuantErrors = mergeQuantErrors(allQuantErrors, max(currentIterationsStats))
    
    #errorbar (numery epok 1-12, średni_błąd, odchylenie_std_błędu, <inne opcje>)
    #plot.errorbar(range(1,13, column2array(quantErrorsStats, 0), column2array(quantErrorsStats, 1), marker = '^', markersize=1, color='green', ecolor='red')
    plot.errorbar(range(1,len(mergedQuantErrors) + 1), column2array(mergedQuantErrors, 0), column2array(mergedQuantErrors, 1), marker = '^', markersize=1, color='green', ecolor='red')
    plot.title("Błędy kwantyzacji w kolejnych iteracjach dla " + str(amount) + " centroidów")
    plot.ylabel('średni błąd kwantyzacji (odchylenie std.)')
    plot.xlabel('numer iteracji') 
    plot.show()
    
    minError = list(avr[0] for avr in quantErrorsStats).index(min(list(avr[0] for avr in quantErrorsStats)))
        
    
    # z tego będzie ładny wykres punktowy centroidów i punktów
    plot.plot(column2array(finalCentroids[minError], 0), column2array(finalCentroids[minError], 1), 'ro', column2array(data, 0), column2array(data, 1), 'gs', markersize=2)
    plot.title("Końcowe położenie centroidów względem punktów dla epoki " + str(minError) + " i " + str(amount) + " centroidów")
    centroid_mark = mlines.Line2D([], [], color='red', marker='o', markersize=10, label='centroid', linewidth = 0)
    point_mark = mlines.Line2D([], [], color='green', marker='s', markersize=10, label='punkt', linewidth = 0)
    plot.legend(handles=[centroid_mark, point_mark])
    plot.ylabel("x")
    plot.xlabel("y") 
    plot.show()

```

```python
plot.bar(['6', '12', '30'], column2array(iterationsStats, 0), color='green', yerr=column2array(iterationsStats, 1))
plot.title("Średnie liczności itercji dla różnych liczności centroidów")
plot.ylabel('średnia liczność iteracji (odchylenie std.)')
plot.xlabel('liczność szukanych centroidów') 
plot.show()
```
